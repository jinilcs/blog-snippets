```python
embeddings_list = []
word_to_index = {}

with open('glove.6B/glove.6B.50d.txt', 'r', encoding='utf-8') as f:

    #Iterate through each line in the file
    for index,line in enumerate(f):

        #Split the line by whitespace to separate the word and its embedding vector
        values = line.split()

        #Extract the word
        word = values[0]

        #Extract the embedding vector as a list of floats
        embedding = list(map(float, values[1:]))

        #Add the word and its corresponding index to the dictionary
        word_to_index[word] = index

        #Add the embedding vector to the list
        embeddings_list.append(embedding)
```

```python
import torch
embeddings_tensor = torch.FloatTensor(embeddings_list)
embedding = nn.Embedding.from_pretrained(embeddings_tensor)
print(embedding)
```

```python
indices = torch.LongTensor([word_to_index['butterfly']])
print(embedding(indices))
```

```python
input_string = "The cat loves to drink milk"
tokens = input_string.lower().split()
indices = torch.LongTensor([word_to_index[token] for token in tokens])
print(embedding(indices).shape)
print(embedding(indices))
```
